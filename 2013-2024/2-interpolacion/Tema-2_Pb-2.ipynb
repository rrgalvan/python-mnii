# Tema 2. Problema 2

Consideremos en el intervalo $[-1,1]$ la función
\begin{equation*}
     f(x) = \frac{1}{1+25x^2}
\end{equation*}
junto a un soporte de $n+1$ puntos equiespaciados en $[-1,1]$.

Representar gráficamente la función junto al polinomio de interpolación de
Lagrange, $p_n(x)$, para $n=\{ 4, 8, 12, 16, 20 \}$. Comprobar gráficamente cómo el error aumenta con $n$.

```python
from __future__ import division, print_function # Sólo si se usa Python 2.x
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = lambda x: 1/(1+25*x*x)

# 1. Gráfica de f
a, b = -1, 1
x = np.linspace(a, b, 100)
plt.axhline(c='k'); plt.axvline(c='k') # Ejes Horizontal y Vertical

# 2. Gráficas de polinomios de interpol.
from scipy.interpolate import lagrange
lista_n = [2, 4, 6, 8, 16, 20] 
lista_colores = ["green", "red", "orange", "pink", "gold", "gray"]

for i in range( len(lista_n) ):
    # a) Plot function f(x)
    plt.plot(x, f(x), "-", lw=4, c='midnightblue', label="$f(x)$")
    # b) Plot nodes
    n = lista_n[i]
    nodos = np.linspace(a, b, n+1)
    plt.plot(nodos, f(nodos), "o", markeredgewidth=1.2, markersize=8, color="gray")
    # c) Plot p_n(x)
    p = lagrange(nodos, f(nodos))
    c = lista_colores[i]
    plt.plot(x, p(x), "-",color=c, lw=3, label="$p_{" + str(lista_n[i]) + "}(x)$")
    # d) Show result
    plt.grid()
    plt.legend()
    plt.show()
```

### Apéndice: Cálculo simbólico de las derivadas de $f$

Sabemos que el error de interpolación está relacionado con el máximo de las derivadas $n$--ésimas. Podemos calcular algunas derivadas utilizando `sympy` (Python simbólico):

```python
import sympy as sp
x = sp.var('x') # Ahora x es una variable simbólica
print("f(x):")
f(x)
```

```python
sp.init_printing() # Impresión mejorada de resultados (usando LaTeX)
print("Derivada primera, f'(x):")
sp.factor(sp.diff(f(x),x))
```

El numerador en la expresión anterior proviene de la derivada del denominador de $f(x)$. Su máximo se alcanza para $x=1$. El mínimo del denominador se tiene para $x=0$. Así tenemos:
$$
M'=\max_{x\in[-1,1]}|f'(x)|=50/1 = 50.
$$

```python
print("Derivada segunda:")
sp.factor(sp.diff(f(x),x,2))
```
